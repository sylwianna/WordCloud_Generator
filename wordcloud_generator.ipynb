```python
import os
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

# # Optional imports
# from docx import Document
# from docx.shared import Inches
```

```python
# Check current working directory
os.getcwd()
```

```python
# Import list of words to include in a word cloud
words_list = open("words_list.txt").read()
words_list
```

```python
# Add words that will be excluded from a cloud
stop_words = [] + list(STOPWORDS)
print(stop_words)
```

```python
# # (Optional) create colours of the same hue but with different luminosities, the hue value is in the range 0-360
# # Example of Google colour picker to set suitable hue: https://www.google.co.uk/search?q=colour+picker
# def random_color_func(word=None, font_size=None, position=None, orientation=None, font_path=None, random_state=None):
#     h = int(360.0 * 21.0 / 255.0)
#     s = int(100.0 * 255.0 / 255.0)
#     l = int(100.0 * float(random_state.randint(60, 120)) / 255.0)

#     return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
# Set wordcloud parameters
wordcloud = WordCloud(font_path = r'C:\Windows\Fonts\Verdana.ttf',
                            stopwords = stop_words,
                            background_color = "rgb(255,255,255)",
                            width=1584, height=396,
                            max_font_size=160, min_font_size=22,
                            colormap=plt.cm.inferno
                            # color_func=random_color_func
                            ).generate(words_list)
```

```python
# Plot a word cloud image preview
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.margins(x=10, y=30)
plt.show()
```

```python
# Save a word cloud image as png file
wordcloud.to_file('wordcloud.png')
```
